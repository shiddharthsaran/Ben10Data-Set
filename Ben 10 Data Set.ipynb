# Ben 10 Data Set

## Downloading the Power Level Video from YouTube using PyTube

```python
from pytube import YouTube
yt=YouTube("https://www.youtube.com/watch?v=5zqOXN6hLAs")
Video_File=yt.streams.filter(progressive=True, file_extension="mp4").order_by("resolution").desc().first().download()
```

```python
print(Video_File)
```

## Importing the necessary Modules

```python
import cv2
import pytesseract 
import pandas as pd
import os
```

## Capturing a Frame for every 8 Seconds 

```python
Video_Capture = cv2.VideoCapture(Video_File)
success,image = Video_Capture.read()
Seconds = 8
FPS = Video_Capture.get(cv2.CAP_PROP_FPS)
Multiplier = FPS * Seconds
Frame_num=0
path="D:\\Data Sets\\Ben 10 Data Set\\Cropped_Images"
while success:
    frameId = int(round(Video_Capture.get(1)))
    success, image = Video_Capture.read()
    if frameId % Multiplier == 0:
        image=image[0:137,0:959]
        cv2.imwrite(os.path.join(path,str(Frame_num)+".png"),image)
        Frame_num+=1
Video_Capture.release()
print ("Complete")
```

## Using OpenCV to read the file and Include a Threshold Value and Using PyTesseract to Convert the Image to String

```python
path="D:\\Data Sets\\Ben 10 Data Set\\Cropped_Images"
w_path="D:\\Data Sets\\Ben 10 Data Set\\Threshold_Images"
Char_and_Pl=list()
for imgn in range(103):
    img=cv2.imread(os.path.join(path,str(imgn)+".png"),0)
    ret,img=cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
    text=pytesseract.image_to_string(img)
    if(len(text)<5):
        cv2.imshow("img",img)
        cv2.waitKey(0)
    else:
        cv2.imwrite(os.path.join(w_path,str(imgn)+".png"),img)
        Char_and_Pl.append(text.split("\n"))
```

## Cleaning the Extracted Data

```python
for data in Char_and_Pl[::-1]:
    if("es | —— -— a ee —————EE a" in data):
        data.remove(data[2])
        data.remove(data[2])
    for index,value in enumerate(data):
        if(value==""):
            data.remove(value)
        if(value=="HUMUNGOUSAUR = 265"):
            data[index]="HUMUNGOUSAUR"
        if(value=="FULL GROWN = 312"):
            data[index]="NORMAL = 265, FULL GROWN = 312"
        if(value=="195"):
            data[index]="NORMAL = 195, GWEN ABSORBED = 750"
        if(value=="GWEN ABSORBED = 750"):
            Char_and_Pl.remove(data)
        if(value=="ULTIMOS" and len(data)==1):
            data.append("280")
        if(value=="DARKSTAR FULL POWER = 990"):
            data[index]="DARKSTAR"
        if(value=="WHEN WEAK = 80 - 160"):
            data[index]="FULL POWER = 990, WHEN WEAK = 80 - 160"
        if(value=="ULT. HUMONGOUSAUR"):
            data[index]="ULTIMATE HUMONGOUSAUR"
        if(value=="NRG IN ARMOR = 140"):
            data[index]="NRG"
        if(value=="PURE ENRGY = 1,160"):
            data[index]="IN ARMOR = 140, PURE ENRGY = 1,160"
        if(value=="OMNIVERSE ©"):
            Char_and_Pl.remove(data)
        if(value=="ULT. ARTIGUANA"):
            data[index]="ULTIMATE ARTIGUANA"
        if(value=="—64133—~*"):
            data[index]="133"
        if(value=="ULT. GRAVATTACK"):
            data[index]="ULTIMATE GRAVATTACK"
        if(value=="TOEPICK = 12.5"):
            data[index]="TOEPICK"
        if(value=="FACE = 1,340"):
            data[index]="NORMAL = 12.5, FACE = 1,340"
        if(value=="EON’S MINIONS"):
            data[index]="EON'S MINIONS"
```

## Splitting the List into Charcters and Power Levels

```python
Character=list()
Power_Level=list()
for data in Char_and_Pl:
    Character.append(data[0])
    Power_Level.append(data[1])
```

## Creating a Series List for all Characters

```python
Ben_10_Series=list()
for i in range(25):
    Ben_10_Series.append("Ben 10: Classic")
for i in range(29):
    Ben_10_Series.append("Ben 10: Alien Force")
for i in range(27):
    Ben_10_Series.append("Ben 10: Ultimate Alien")
for i in range(16):
    Ben_10_Series.append("Ben 10: Omniverse")
```

## Creating a DataFrame with Characters,Power Levels and Universe

```python
import pandas as pd
BTDF=pd.DataFrame(list(zip(Character,Power_Level,Ben_10_Series)),columns=["Character","Power_Level","Ben_10_Series"])
```
